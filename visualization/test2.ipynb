```python

import pandas as pd
import matplotlib.pyplot as plt

plt.title('shopData')
plt.rcParams['font.sans-serif'] = ['SimHei']
shopData = pd.DataFrame(pd.read_csv(r'shopsData.csv', nrows=10))
part = shopData[['name', 'quality', 'environment', 'service']]
name = list(part['name'])
quality = list(part['quality'])
environment = list(part['environment'])
service = list(part['service'])
plt.xticks(rotation=90)
plt.plot(name, quality, marker='o')
plt.plot(name, environment, marker='*')
plt.plot(name, service, marker='d')
plt.legend(['quality', 'environment', 'service'])
plt.show()
```
