```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('tips.csv')
print(data)
fig,ax = plt.subplots(2,1,figsize=(10,10))
sns.barplot(x='day',y= 'total_bill',data = data , hue= 'smoker',palette='husl',order=['Thur', 'Fri','Sat','Sun'],ax=ax[0])
sns.barplot(x='day',y= 'total_bill',data = data , hue= 'sex',palette='husl',order=['Thur', 'Fri','Sat','Sun'],ax=ax[1])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('titanic.csv')
print(data)
survived = data[data.survived == 1]
fig,ax = plt.subplots(2,1,figsize=(10,10))
sns.countplot(x='class',data = data , hue= 'who',palette='husl',order=['First', 'Second','Third'],ax=ax[0])
sns.barplot(x='class',y= 'survived',data = data , hue= 'who',palette='husl',order=['First', 'Second','Third'],ax=ax[1])
plt.show()
```

```python
from pyecharts.charts import Map,Bar
from pyecharts import options as opts
from pyecharts.globals import ChartType, SymbolType
import pandas as pd
```

```python
shopData = {'Name':['公司A','公司B','公司C','公司D','公司E'],
           '商品A':[98,78,97,70,72],
           '商品B':[93,91,80,83,73],
            'Location':['辽宁省','河北省','黑龙江省','甘肃省','江苏省']
           }
data = pd.DataFrame(shopData)
data
data['sales'] = data['商品A'] + data['商品B']
data['Location'] = data.Location.apply(lambda x:x.replace('省',''))
dataList = data[['Location','sales']].groupby(by=['Location'],sort=False).sum().sort_values(by=['sales'],ascending=False)
dataList = [z for z in dataList.sales.items()]
```

```python
map = Map()
map.add("浙江省商品销量分布情况", dataList, "china")
map.set_global_opts(
            title_opts=opts.TitleOpts(title="商品在全国总销量"),
            visualmap_opts=opts.VisualMapOpts(max_=200,min_=140, is_piecewise=True),
        )
map.render_notebook()
```
