```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('seeds.data',sep= ',',header = None)
dataArr = np.array(file)
data = pd.DataFrame(dataArr,columns=['a','b','c','d','e','f','g','ttype'])
print(data)
```

```python
sns.pairplot(data,vars= ['a','b','c','d','e','f','g'],diag_kind= 'kde',hue ='ttype')
```

```python
sns.pairplot(data,vars= ['a','b','c'],diag_kind= 'kde',hue ='ttype',kind= 'reg')
```

```python
sns.PairGrid(data,vars =['a','d','e'],hue="ttype").map_diag(plt.hist).map_offdiag(sns.kdeplot)
# map_diag(plt.hist)#对角线单变量子图
# map_offdiag(sns.kdeplot)#非对角线
```
