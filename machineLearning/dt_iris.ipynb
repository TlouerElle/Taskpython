```python
# 导入包
from sklearn import datasets,tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score, recall_score,classification_report
import matplotlib.pyplot as plt
```

```python
# 载入数据集并划分
iris_data = datasets.load_iris()
x_train,x_test,y_train,y_test = train_test_split(iris_data.data,iris_data.target,test_size = 0.30,random_state = 20)
# x_train    划分出的训练集数据（返回值）
# x_test    划分出的测试集数据（返回值）
# y_train    划分出的训练集标签（返回值）
# y_test    划分出的测试集标签（返回值
#花萼长度、花萼宽度、花瓣长度、花瓣宽度（花萼宽度、花瓣长度）

#目标属性 三种类别山鸢尾花、杂色鸢尾花、维吉尼亚鸢尾花
label_list = ['mountainIris', 'variegatedIris', 'virginiaIris']
```

```python
# 训练并测试画图
#'gini'对应CART算法，'entropy'对应ID3算法
dt_model = tree.DecisionTreeClassifier(criterion='gini', min_samples_leaf=3)
dt_model.fit(x_train, y_train)
%matplotlib inline
fig=plt.figure(figsize=(10,10))
tree.plot_tree(dt_model,filled="True",feature_names=["Sepal length","Sepal width","Petal length","Petal width"],class_names=['mountainIris', 'variegatedIris', 'virginiaIris'])
plt.savefig(fname="tree.jpg")
y_test_hat = dt_model.predict(x_test)
print('accuracy score : ',end = '')
print(accuracy_score(y_test, y_test_hat))
print('precision score : ',end = '')
print(precision_score(y_test, y_test_hat,average='macro'))
print('recall score : ',end = '')
print(recall_score(y_test, y_test_hat,average='macro'))
target_names = ['mountainIris', 'variegatedIris', 'virginiaIris']
print(classification_report(y_test, y_test_hat))
```
